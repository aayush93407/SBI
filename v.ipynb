```python
import sys
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
import numpy as np
import cv2
from PyQt5.QtWidgets import QApplication, QFileDialog

# Define model architecture (must match training setup)
def build_model():
    base_model = EfficientNetB0(weights=None, include_top=False, input_shape=(224, 224, 3))
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(128, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)  # Adjust activation if multi-class
    model = Model(inputs=base_model.input, outputs=output)
    return model

# Initialize model
model = build_model()

# Restore weights from checkpoint
checkpoint_path = "C:/Users/AAYUSH THE GREAT/OneDrive/Desktop/vehicle/"
checkpoint = tf.train.Checkpoint(model=model)
checkpoint.restore(tf.train.latest_checkpoint(checkpoint_path)).expect_partial()

# Function to preprocess an input image
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.resize(image, (224, 224))
    image = image / 255.0  # Normalize pixel values
    image = np.expand_dims(image, axis=0)  # Add batch dimension
    return image

# Function to make predictions
def predict_vehicle(image_path):
    image = preprocess_image(image_path)
    prediction = model.predict(image, verbose=0)
    return "Vehicle Detected" if prediction[0][0] > 0.45 else "No Vehicle Detected"

# Function to open file dialog and select an image
def select_image():
    app = QApplication(sys.argv)  # Initialize PyQt application
    file_path, _ = QFileDialog.getOpenFileName(None, "Select an Image", "", "Image Files (*.png *.jpg *.jpeg *.bmp *.gif)")
    return file_path

# Select image dynamically
image_path = select_image()

# Perform prediction if an image is selected
if image_path:
    print("Selected Image:", image_path)
    result = predict_vehicle(image_path)
    print("Prediction:", result)
else:
    print("No image selected.")
```
